Algumas informações são desncessárias e outras podem ser alocadas em uma única tabela

```python
dados_demand['Ciclo Lento'] = dados_datas.loc[:,'CICLO_LENTO']
dados_demand = dados_demand.drop('FABRICA',axis=1)
dados_demand
```

```python
dados_trans['Gruas'] = dados_gruas['QTD_GRUAS']
dados_trans['%'] = dados_gruas['PORCENTAGEM_VEICULOS_MIN']
dados_trans = dados_trans.drop('DIA',axis=1)
dados_trans
```

```python
colunas_interessadas = dados_BD_UP.columns[:5]
dados_BD_UP = dados_BD_UP[colunas_interessadas]
dados_BD_UP
```
